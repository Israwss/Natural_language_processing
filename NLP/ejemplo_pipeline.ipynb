```python
from transformers import pipeline


classifier = pipeline('sentiment-analysis',
                      model="pysentimiento/robertuito-sentiment-analysis")

sentences = ["¡Me encanta el macroentrenamiento, los cursos y tutoriales son lo máximo!",
                "Odio los lunes, no sirven para nada.",
                "El artículo es medio bueno; me gustó como presentaron los resultados."]

results = classifier(sentences)
for result in results:
    print(f"polaridad: {result['label']}, score: {round(result['score'], 4)}")
```
